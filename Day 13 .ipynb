```python
# Tic Tac Toe Game

def print_board(board):
    print()
    for row in board:
        print("  | ".join(row))
        print(" -"*6)
    print()
        
        
def check_winner(board, player):

    #check row winner 
    for row in board:
        if all([cell == player for cell in row]):
            return True

    
    #Check Col winner 
    for col in range(3):
        
        #checking all the rows for each columns
        if all([board[row][col] == player for row in range(3)]):
            return True
    
    
    #checking diagonal conditions
    if all([board[i][i] == player for i in range(3)]) or all([board[i][2-i] == player for i in range(3)]):
            # Left to right Diagonal                     # Right to left Diagonal 
        return True

    
    #Function's return false : No one won
    return False
        
    
    
    
def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    current_player= "X"
    
    for turn in range(9):
        print_board(board)
        
        row = int(input(f"Player {current_player}, enter row (1-3):"))

        #Substracting 1 from Input of user to make it index 
        row = row-1
        
        col = int(input(f"Player {current_player}, enter Col (1-3):"))
        
        #Substracting 1 from Input of user to make it index 
        col = col -1
        
    
        if board[row][col] == " ":
            board[row][col] = current_player
            
            if check_winner(board,current_player):
                print_board(board)
                print(f"Player {current_player} wins!!")
                return 
            
            current_player = "O" if current_player == "X" else "X"
            
            
            # do this -> current_player = "O" only if current_player == "X"
            
            # if current_player is not = "X" (i.e current_player == 'O')then change the current player to x 
            
        else: print("Cell is already taken. Try again")
                
                
    print_board(board)
    print("It's a Draw !!")
        
        
        
        
        
tic_tac_toe()
```
