```python
#Hangman Game 

import random

def hangman():
    words = ["python", "hangman", "challenge", "Yash"]
    word_to_guess = random.choice(words)
    #print(f"=== Hidden : {word_to_guess}")
    
    Apun_Ka_Guess = ["_"]*len(word_to_guess)
    attempts = 6 
    
    #Creating empty set 
    guessed_letter = set()
    
    print("-- Hello, welcome to Hangman -- ")
    
    while attempts > 0 and "_" in Apun_Ka_Guess:
        print(f"\nWord: {' '.join(Apun_Ka_Guess)}")
        print(f"Attempts remaining: {attempts}")
        
        guess = input("Guess a letter: ").lower()
    
        if guess in guessed_letter:
            print("You have already Guessed that. ")
        elif guess in word_to_guess:
            guessed_letter.add(guess)
            
            for i, letter in enumerate(word_to_guess):
                if letter == guess:
                    Apun_Ka_Guess[i] = guess
        else:
            guessed_letter.add(guess)
            attempts-=1
            print("Incorrect Guess")
    
    
    if "_" not in Apun_Ka_Guess:
        print(f"Congrats. You passed in {attempts} by guess the word {''.join(Apun_Ka_Guess)}")
    else:
        print(f"Game Over! the word was : {word_to_guess}")

        
        
        
        
hangman()
```

```python
#Max In List using 100% OOPS
#Time Complexity : O(N)
#Space : O(1)


class MaxFinder:
    def __init__(self, lst):
        self._lst = lst
        self._max_value = None
        
        # Start automation here....
        self._initialize_max_value()
        
    def _initialize_max_value(self):
        if not isinstance(self._lst, list):
            raise TypeError("Input should be of type: List")
        
        if not self._lst:
            raise ValueError("List cannot be empty")
        
        # Changed: Move self._max_value assignment out of unreachable code block
        self._max_value = self._lst[0]
        return self
    
    def find_max(self):
        try:
            for num in self._lst:
                # calling _update_max() method here from find_max() method
                self._update_max(num)
        except TypeError:
            print("Type Error: Input should be a List")
        return self
    
    def _update_max(self, num):
        try:
            # Changed: Added type checking for the num parameter
            if not isinstance(num, (int, float)):
                raise TypeError(f"Invalid element {num} in the list")
            # Changed: Correct the logic to properly update max value
            if self._max_value is None or num > self._max_value:
                self._max_value = num
        except TypeError as e:
            print(e)
            
        return self
    
    def get_max(self):
        if self._max_value is None:
            raise ValueError("Max value not found.")
        return self._max_value

# Child Class
class AdvancedMaxFinder(MaxFinder):
    def __init__(self, lst, ignore_negatives=False):
        self.ignore_negatives = ignore_negatives
        
        # Inheritance 
        super().__init__(lst)
        
    def _initialize_max_value(self):
        if not isinstance(self._lst, list):
            raise TypeError("Input should be of type: List")
        
        if not self._lst:
            raise ValueError("List cannot be empty")
        
        # Changed: Corrected _max_value initialization logic
        self._max_value = float("-inf") if self.ignore_negatives else self._lst[0]
        return self
    
    # Function overriding
    def find_max(self):
        try:
            for num in self._lst:
                if self.ignore_negatives and num < 0:
                    continue 
                self._update_max(num)
        except TypeError:
            print("Type Error: Input should be a List")
        return self

    def _update_max(self, num):
        try:
            super()._update_max(num)
        except TypeError:
            print(f"Type Error: Invalid element {num} in the list")

# Test the functionality
try:
    lst1 = [3, 1, 4, -1, 5, 559, 2, 6, -5]
    lst2 = [-10, -20, -30, -40, -50, 10, 20, 30]

    max_finder1 = AdvancedMaxFinder(lst1, ignore_negatives=True).find_max()
    max_finder2 = AdvancedMaxFinder(lst2, ignore_negatives=True).find_max()

    print(f"The max element in list1 (ignore_negatives=True): {max_finder1.get_max()}")
    print(f"The max element in list2 (ignore_negatives=True): {max_finder2.get_max()}")

    max_finder1 = AdvancedMaxFinder(lst1, ignore_negatives=False).find_max()
    max_finder2 = AdvancedMaxFinder(lst2, ignore_negatives=False).find_max()

    print(f"The max element in list1 (ignore_negatives=False): {max_finder1.get_max()}")
    print(f"The max element in list2 (ignore_negatives=False): {max_finder2.get_max()}")

except ValueError as ve:
    print(f"ValueError: {ve}")
except TypeError as te:
    print(f"TypeError: {te}")
```
