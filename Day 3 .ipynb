```python
#Modified Rev Number prog. 
#Time - O(n)
#Space - O(1)

def ReverseTheNumber(n):
    
    Flag_Of_Number_Is_Negative = n < 0       # a flag which tells me if the number is negative. 
    n = abs(n)
    
    Final_Reversed_Number = 0
    
    while n > 0 :
        rem = n%10
        Final_Reversed_Number = (Final_Reversed_Number * 10)+rem
        n = n//10

    if Flag_Of_Number_Is_Negative == True: 
        Final_Reversed_Number = -Final_Reversed_Number # saving the same number as negative
        
    return Final_Reversed_Number


Negative_Ans = ReverseTheNumber(-12345)
Positive_Ans = ReverseTheNumber(43928)


print(f"Rev. of Negative number: {Negative_Ans}")
print(f"Rev. of Positive number: {Positive_Ans}")


#checking if the reversed number is still of type number. hence multiplying it by 100 just to check. 
print(Negative_Ans*100)
```

```python
#FizzBuzz : Print all numbers from 1 to 100 but not, 
# If Divisible by 3 -> Print Fizz
# If Divisible by 5 -> Print Buzz
# If Divisible by 3&5 -> Print FizzBuzz
#Time - O(n)
#Space - O(1)



def fizzbuzz(n):
                     #endPt
    for i in range(1,n+1):
        if i % 3 == 0 and i%5==0:
            print("FizzBuzz")
        elif i%3==0:
            print("Fizz")
        elif i%5==0:
            print("Buzzz")
        else:
            print(i)
            
fizzbuzz(100)
```
