```python
#Passing The Parcel 


from collections import deque


def Game(players, num):
    
    queue = deque(players)
    
    while len(queue) > 1: 
        
        for i in range(num):
            queue.append(queue.popleft())
            
        #Removal of the Guy after N passes in the above loop 
        print(f"Eliminating {queue[0]}")
        queue.popleft()
        
        
    return queue[0]


players = ["Kartik", "Darshan", "Balraj", "Priyansh", "Pranav"]
            
num = 7 


winner = Game(players, num)

print(f"Winner is .... {winner}")
```

```python
class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []
        
    def add_child(self, child_node):
        self.children.append(child_node)
        

class TreeMaze:
    def __init__(self, root_node):
        self.root = root_node
    
    def display_rooms(self, node=None, level=0):
        if node is None:
            node = self.root
        
        print(" " * level + node.name)
        
        for child in node.children:
            self.display_rooms(child, level + 1)
    
    def find_path(self, start_node, target_name, path=None):
        if path is None:
            path = []
        
        path.append(start_node.name)
        
        if start_node.name == target_name:
            return path
        
        for child in start_node.children:
            result = self.find_path(child, target_name, path.copy())
            if result:
                return result
        
        return None


root = TreeNode("Start")
node_a = TreeNode("A")
node_b = TreeNode("B")
node_c = TreeNode("C")
node_d = TreeNode("D")
node_e = TreeNode("E")
node_f = TreeNode("F")
node_g = TreeNode("G")


root.add_child(node_a)
root.add_child(node_b)
node_a.add_child(node_c)
node_b.add_child(node_d)
node_d.add_child(node_e)
node_b.add_child(node_f)
node_b.add_child(node_g)

maze = TreeMaze(root)

print("--- Maze: ---")
maze.display_rooms()

target = "E"
path_to_E = maze.find_path(root, target)

if path_to_E:
    print(f"\nPath to {target}: {' -> '.join(path_to_E)}")
else:
    print(f"\n{target} not found")
```
