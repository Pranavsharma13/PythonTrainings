```python
# Balance of Paranthesis - > Stack Implementation 


class ApunKaStack:
    def __init__(self):
        self.item = []
    
    
    def is_empty(self):
        return len(self.item) == 0
    
    
    def push(self,New_Data):
        self.item.append(New_Data)
    
    
    def pop(self):
         if not self.is_empty():
            return self.item.pop()
         return None 
    
    def peek(self):
        if not is_empty():
            return self.item[-1]
        return None 
    
    
def is_balanced(parentheses_to_check):
    
    MyStack = ApunKaStack()
    reference_parentheses = {')': '(', ']': '[', "}": "{"}
    
        
    for char in parentheses_to_check:
        if char in '([{' : #opening Brackets 
            MyStack.push(char)
        
        elif char in ']})': # Closing bracket 
            
            if MyStack.is_empty() or MyStack.pop() != reference_parentheses[char]:
            # if stack is empty or char in the stack doesn't match with the reference 
                return False 
            
    return MyStack.is_empty()    
        
        
        
expression = "{ { [() ()] } }"
print(f"Expression {expression} is Balanced ? :- {is_balanced(expression)}")


expression2 = "{ [ ( ] ) }"
print(f"Expression {expression2} is Balanced ? :- {is_balanced(expression2)}")        
```

```python
#Round Robin Prog using Singly Circular Linked List 

class Node: 
    def __init__(self,data):
        self.data = data 
        self.next = None 


        
class CircularLinkedList:
    def __init__(self):
        self.head = None 
        self.tail = None 
        
        
    def append(self,data):
        
        new_node = Node(data)

        #if list is empty
        if not self.head:
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
        
        else:
            self.tail.next = new_node
            self.tail = new_node
            self.tail.next = self.head
            
        
    def display(self):
        nodes = []
    
        if self.head:
            current = self.head
            
            while True:
                nodes.append(str(current.data))
                current = current.next 
                
                #Used to check if current is head. 
                #if yes break the loop as we are at the start of the list and we are done traversing 
                if current == self.head:
                    break
                
                
        print(" -> ".join(nodes)+ " -> ...(Circular)")        
            
        
        
    def round_robin_schedule(self, time_quantum):
        
        if not self.head:
            return 
        
        
        current = self.head
        
        while True: 
            print(f"Process {current.data} is running for {time_quantum} ms\n")
            
            current = current.next
            
            if current == self.head:
                break
    
        
        
        
        
process = CircularLinkedList()

process.append("Process 1 ")
process.append("Process 2 ")
process.append("Process 3 ")
process.append("Process 4 ")
process.append("Process 5 ")
process.append("Process 6 ")
        
        
print(" --- List of Processes using Circular Linked List --- \n")        
process.display()

print(" --- Round Robin Scheduling Algo is working with Time Slice of 100ms --- \n")
process.round_robin_schedule(100)
        


```

```python
# Dice Roll Game 

import random 


class Dice: 
    def __init__(self):
        self.value = 1
        
        
    def roll(self):
        self.value = random.randint(1,6)
        
    
    def get_value(self):
        return self.value
    

    
class Roll_The_Dice:
    def __init__(self):
        self.dice1 = Dice() # Objects Dice 1 and 2 are created for Dice class 
        self.dice2 = Dice()
        
        
    def roll_it(self):
        self.dice1.roll()
        self.dice2.roll()
        
        return self.dice1.get_value(), self.dice2.get_value() 
        


    def display_the_numbers(self):
        value1, value2 = self.roll_it()
        print(f"Dice 1: {value1}")
        print(f"Dice 2: {value2}")
        
        
        
start = Roll_The_Dice()
start.display_the_numbers()
        
```
